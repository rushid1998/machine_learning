```python
import numpy as np

import pandas as pd
import difflib
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#data collection
```

```python
data = pd.read_csv("/content/tmdb_5000_movies.csv")
```

```python
data.head(5)
```

```python
data.shape
```

```python
#selecting relevent feature for recommended
```

```python
data.columns
```

```python
select_features=["keywords","genres","tagline","tagline","overview"]
print(select_features)
```

```python
for feature in select_features:
  data[feature]=data[feature].fillna('')
```

```python
combined_feature =data["genres"]+' '+ data["keywords"]+' '+ data["title"]+' '+ data["tagline"]+' '+ data["overview"]
print(combined_feature)
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combined_feature)

print(feature_vector)
```

```python
#cosine similarity
```

```python
similarity=cosine_similarity(feature_vector)
print(similarity)
```

```python
print(similarity.shape)
```

getting movie name from the user

```python
movie_name=input("enter the favorite movie name:")
```

```python
title_list= data["title"].tolist()
print(title_list)
```

```python
find_close_match=difflib.get_close_matches(movie_name,title_list)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
data = data.reset_index(drop=True)
index_of_movie = data[data['title'] == close_match].index[0]
print(index_of_movie)
```

```python
data.head()
```

```python
#getting a list of similar movie
similarity_score= list(enumerate(similarity[index_of_movie]))
print(similarity_score)
len(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score,key = lambda x:x[1],reverse= True)
print(sorted_similar_movies)
```

```python
print("movies suggested for you: \n")
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=data[data.index == index]['title'].values[0]
  if(i<10):

    print(i,'-',title_from_index)
    i+=1
```
